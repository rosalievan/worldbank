```python
x= data['Report Year'].unique()
plt.figure(figsize=(10, 6)) 

for item in data['Region'].unique():
    y = data.loc[data['Region'] == item].groupby(['Report Year'])['WBL INDEX'].mean()
    plt.plot(x, y, label = item)
plt.legend()
plt.xlabel("Year")
plt.ylabel("WBL Index")
plt.ylim([0, 100])
plt.show()
```

```python
data = data.replace('Yes', 1)
```

```python
data = data.replace('No', 0)
```

```python
categories = ['ASSETS', 'MOBILITY', 'PENSION', 'ENTREPRENEURSHIP', 'PARENTHOOD', 'MARRIAGE', 'PAY', 'WORKPLACE']

dict = {}
for year in data['Report Year'].unique():
    dict[year] = 0
```

```python
for country in data['Economy'].unique():
    for category in categories:
        temp = data.loc[data['Economy'] == country]
        temp = temp.merge(temp[[category]].diff())
        print(temp)
        year = temp.iloc['Report Year']
        dict[year] += 1
```
